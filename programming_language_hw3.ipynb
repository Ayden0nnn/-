```python
#抓取博客來原始碼網頁
import urllib.request as req
import bs4 #beautifulsoup4套件
import json

url = "https://www.ptt.cc/bbs/Coffee/index.html"

#建立 Request物件，附加 Request Headers 的資訊
request = req.Request(url, headers = {
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"
})

with req.urlopen(request) as response:
    data = response.read().decode("utf_8")
print(data)
```

```python
root = bs4.BeautifulSoup(data, "html.parser")
root.title.string
```

```python
title=root.find("div", class_="title") #尋找 class_="txt" 的 div 標籤 #find 會找到期中一個符合條件的標籤
title
```

```python
titles=root.find_all("div", class_="title")
for txt in titles:
    if txt.a != None: #應該 None不會印出來才對......><
        print(txt.a.string)
```

```python
import json
from google.colab import files

titles = root.find_all("div", class_="title")
titles_text = [txt.a.string for txt in titles if txt.a is not None]

# 構建 JSON 對象列表
titles_list = []
for i, title_text in enumerate(titles_text):
    # title_characters = list(title_text)

    # 構建 JSON 對象
    title_json = {
        "index": i + 1,  # 索引
        "title_text": title_text,  # 原始標題文字

    }

    titles_list.append(title_json)

# 存儲路徑
store = '/content/all_titles.json'

# 將 JSON 對象列表寫入檔案
with open(store, 'w') as json_obj:
    json.dump(titles_list, json_obj, ensure_ascii=False)

# 下載檔案
files.download(store)

print("儲存成功")
```
